#### Exercise 1) Normal or Uniform?


```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

#### a) Find the range of Y for each hypothesis. What should the decision be if an observation follows |y| > 1
We can see that the uniform distribution can only take values between -1 and 1, so a |y| > 1 would be impossible under this hypothesis.
The normal distribution has infinite range, with the probability of a value far in the tails being extremely low.

#### b) Plot the 2 hypothesis as a function of y

```python
# data for plotting
N = 1000 # number of samples
h0_data = np.random.uniform(-1, 1, N)
h1_data = np.random.normal(0, 1, N)

# calculate posterior probability of h0 and h1
h0_prior = 0.4
h1_prior = 1 - h0_prior

h0_likelihood = stats.uniform.pdf(h0_data, -1, 2)
h1_likelihood = stats.norm.pdf(h1_data, 0, 1)

h0_posterior = h0_likelihood * h0_prior
h1_posterior = h1_likelihood * h1_prior

# plot the data
plt.figure()
plt.plot(h0_data, h0_posterior, 'r.', label='h0') # red dots (like a scatter plot)
plt.plot(h1_data, h1_posterior, 'b.', label='h1')
plt.xlabel('Data')
plt.ylabel('Posterior Probability')
plt.legend()
plt.title(f'priors: h0={h0_prior}, h1={h1_prior}')
plt.show()
```

We can observe that for h0 priors > 0.45 the MAP desicion region for H0 is [-1, 1].

With a bit lower around 0.45 > y > 0.35 the MAP region for H0 is only just around the values -1 and 1.

For h0 with a prior lower than 0.35 the MAP region is 0
